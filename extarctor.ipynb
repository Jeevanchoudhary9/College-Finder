```python
# WHole Code

import pandas as pd
import pdfplumber

# Open the PDF
with pdfplumber.open("DSE_CAP1_CutOff_2023_24.pdf") as pdf:
    combined_tables = []  # List to store the combined tables
    
    for page in pdf.pages:
        # Extract tables
        raw_tables = page.extract_tables()
        temp = []  # Temporary storage for grouping related rows
        
        for table in raw_tables:
            if len(table) == 2 and "Choice Code" in table[1][0]:  # Detect "header" rows
                if temp:  # If temp already has a grouped table, save it
                    combined_tables.append(temp)
                temp = []  # Reset for a new group
                temp.extend(table)  # Add the header table
            else:
                temp.extend(table)  # Append the current table rows to the group
        
        if temp:  # Add the last grouped table from this page
            combined_tables.append(temp)

# Display the combined tables
for i, table in enumerate(combined_tables):
    print(f"\nCombined Table {i + 1}:\n")
    for row in table:
        print(row)


all=[]
for table in combined_tables:
    original={"College Name" :table[0][0],table[1][0]:table[1][1],table[1][2]:table[1][3]}
    keys=table[2][1:]
    values=table[3][1:]
    values = [
    value.split('(')[1].split(')')[0].replace('%', '') 
    if '(' in value and ')' in value else None 
    for value in values
    ]
    result = dict(zip(keys, values))
    original.update(result)
    all.append(original)

df = pd.DataFrame(columns=['College Name', 'Choice Code :','Course Name :'])
# if 'Open' not in df.columns:
#     df['Open'] = None  # Or you can initialize with other default values if needed

new_row = pd.DataFrame(all)
df = pd.concat([df, new_row], ignore_index=True)
df.to_excel('output.xlsx', index=False)
df.to_csv('data.csv')
```
